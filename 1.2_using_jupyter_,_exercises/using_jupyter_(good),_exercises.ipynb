```python
squares = [i**2 for i in range(10)]

squares2 = list()
for i in range(10):
    squares2.append(i**2)
    
print(squares)
print(squares2)

```

```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares2 = list()
for i in range(10):
    squares2.append(i**2)
```

3.66 and 4.04 $\mu$s is the average time per code execution. 105 and 140 ns is the standard deviation of these multiple runs (the smaller the number, the higher the consistancy of timings). 1 run is 100000 loop (loop = one code execution).

This means that **squares**, on average, generates faster with a higher consistancy, compared to **squares2**.
