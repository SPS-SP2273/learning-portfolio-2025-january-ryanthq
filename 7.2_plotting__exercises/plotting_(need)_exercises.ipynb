## Exercise 1

```python
import numpy as np
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

## Exercise 2

```python
x_list = list(range(0,51))
x_array = np.array(x_list)
y_array = np.exp(-x_array / 10) * np.sin(x_array)

plt.plot(x_array, y_array)
plt.title(r"$\mathbf{y = e^{-x/10}sin(x)}$", fontsize=14)
plt.xlabel(r"$\mathbf{x}$", fontsize=12, color='orange')
plt.ylabel(r"$\mathbf{y}$", fontsize=12, color='orange', rotation=0)
plt.grid(alpha=0.25)

plt.show()
```
